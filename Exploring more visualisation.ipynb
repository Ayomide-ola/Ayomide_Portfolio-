## Bar plot using pandas and seaborn

```python
df['cut'].unique()
```

### Sampling the data

```python
cut_count_table =pd.crosstab(index= df['cut'],columns='count')
```

```python
cut_count_table
```

```python
cut_count_table.plot(kind ='bar')
```

```python
sns.catplot(data=df, x='cut', kind='count');
```

```python
sns.catplot(data=df, y='cut', kind='count');
```

```python
#sns.catplot(data=df, y='cut', kind='count', palette ='green');
```

```python
sns.catplot(data=df, y='cut', kind='count', palette ='Set2', aspect=1.5);
```

### Mean price for each category 

```python
sns.barplot(data=df, x='cut', y= 'price', palette ='Set2', estimator= 'mean');
```

```python
sns.barplot(data=df, x='cut', y= 'price', palette ='Set2', estimator= 'mean',errorbar=('ci', 68));
```

```python
sns.barplot(data=df, x='cut', y= 'price', palette ='Set2', estimator= 'median',errorbar=('ci', 95));
```

## Rearranging the categories 

```python
sns.barplot(data=df, x='cut', y= 'price', palette ='Set2', estimator= 'mean',
            errorbar=('ci', 95),order=['Ideal','Good','Very Good','Fair','Premium']);
```

```python
sns.barplot(data=df, x='cut', y= 'price',hue='color')
plt.legend(loc= 'upper right', bbox_to_anchor= (1.2,1)); #What is the diffrence between the price based on the cut and the colour
```

```python
ax = sns.barplot(data=df, x='cut', y= 'price',hue='color')
ax.legend(loc='upper right',ncol=4)
plt.legend(loc= 'upper right', bbox_to_anchor= (1.2,1)); 
```

```python
ax = sns.barplot(data=df, x='cut', y= 'price',hue='color')
ax.legend(loc='upper right',ncol=4)
ax.set_xlabel('Cut',fontdict= {'fontsize':15})
ax.set_ylabel('Price',fontdict= {'fontsize':15})
plt.legend(loc= 'upper right', bbox_to_anchor= (1.2,1));
```

```python
ax = sns.barplot(data=df, x='cut', y= 'price',hue='color')
ax.legend(loc='upper right',ncol=4)
ax.set_xlabel('Cut',fontdict= {'fontsize':15})
ax.set_ylabel('Price',fontdict= {'fontsize':15})
ax.set_xticklabels(ax.get_xticklabels(),fontsize=13,rotation=30)
plt.legend(loc= 'upper right', bbox_to_anchor= (1.2,1));
```
