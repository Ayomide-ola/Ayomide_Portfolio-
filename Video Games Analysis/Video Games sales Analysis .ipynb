Five number summary means summarising data under five heads namely: Mean, median, Mode, Min and Max. 
1. ***Mean:*** This is the average of the observations in a column. It is calculated by summing together all observations in a variable and dividing the sum by the count of all observation. In 'df' the mean of any variable would be the sum of all variables divided by 16179. It is the most common indicator of central tendency (Torres-Reyna, 2008)
2. ***Median:*** This is the middle or average of the middle observations of a variable after the observations have been sorted in ascending order, i.e from the lowest observation to the highest. This in our number summary would be the 50% of the variable. It is another measure of central tendency (Torres-Reyna, 2008).
3. ***Mode:*** Mode refers to the most frequent observation in a given variable. 
4. ***Min:*** This is simply, the minimum value in each variable. 
5. ***Max:*** This is the maximum value in each variable. 
With this five statistical operations, we can summarise the locations of observations in each variable.
##### Variability
In addition to knowing the locations of our observations, it is important to explore the variability of the observations from the mean (50%). To do this we would have to calculate: Variance, range, and Standard deviation. 

6. ***Variance:**** Variance, as the name implies, is the measure of dispersion of data from the mean (50%) (Torres-Reyna, 2008). This is calculated by squaring the difference between mean and each observation in a column i.e (each column observation minus mean of the column), and dividing the summation of squares by the total number of observation minus 1 (Hayes, 2023).
7. ***Standard deviation:*** Standard deviation is the square root of variance. Variables with high standard deviation relative to the mean have their data spread out from the mean, while those with small standard deviation have their data tightly grouped close to the mean.

# Part: Hypothesis testing
I want to conduct a hypothesis to see if there is difference between the means of the three groups ['Genre', 'EU_sales', 'JP_sales'] in the numerical variable. when alpha = 0.05

```python
Games.head()
```

```python
from scipy.stats import f_oneway
```

```python
# Extract the data for each group
Na_sales = Games['NA_Sales']
Eu_sales = Games['EU_Sales']
Jp_sales = Games['JP_Sales']
```

```python
f_statistic, p_value = f_oneway(Na_sales, Eu_sales,Jp_sales)
print("F-statistic:", f_statistic)
print("P-value:", p_value)
```

## Meaning of result
1. F-statistic: This  measures the ratio of the variance of between group to within group. This means that it compare the difference between the variance of each sales (within group) and the variance on a sale compareded to the other, e.g. how is EU sale diffrent fro jp sale (between group). A higher F-statistic suggests a larger difference between the group means relative to the variability within each group.
2.  p-valu: Is used as a bases to reject of accept an hypothesis. The p value in this contexte iy small, indicating strong evidence against the null hypothesis.
## Summary:
Generally if the pvalue is lesser than alpha we reject the null hypotesis. In this case the p value is not lesser than the alpha, so , we reject the null hypothesisThis means that there is  significant evidence to conclude that there is a difference between the means of ('NA_sales', 'EU_sales', 'JP_sales') in the games dataset.ps.
