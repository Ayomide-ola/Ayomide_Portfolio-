```python
# Import packages
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data
df = sns.load_dataset('penguins').rename(columns={'sex': 'gender'})
df.head()
```

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender');# alpha controls transperency of the data points
```

```python
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm',  hue='species', size='gender');
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender');
```

```python
# Plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1));
```

```python
# Change default style
sns.set_style('whitegrid') # This gives the grid..
# Plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1));
```

```python
# Change default context
sns.set_context('talk') # This gives the grid..
# Plot
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.4, 1));
```

```python
sns.set_palette('rainbow')# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1));
```

```python
# Change default palette
sns.set_palette(['green', 'purple', 'red'])
# Plot
plt.figure(figsize=(9, 5))
sns.scatterplot(data=df, x='body_mass_g', y='bill_length_mm', alpha=0.7, hue='species', size='gender')
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1));
```
