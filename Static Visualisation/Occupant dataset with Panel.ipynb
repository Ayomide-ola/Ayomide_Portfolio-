```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_agg import FigureCanvas 
```

```python
%matplotlib inline
```

```python
def mpl_plot(avg, highlight):# Creating a function that works for all
    fig = Figure() # Create a vizualisation
    FigureCanvas(fig) # Create a canves 
    ax = fig.add_subplot() # Create a subplot 
    avg.plot(ax=ax) # This plots line plot, scatter for example will be avg. scatter
    if len(highlight): highlight.plot(style='o', ax=ax)  #Plot if I have outliers, and if there are no outliers don't plot
    return fig 
def find_outliers(variable = 'Temperature', window = 30, sigma = 10, view_fn = mpl_plot):# variable, the time window for exponential, sigma= 0.3(normally): sigma value is what controls the outer range 
    avg = data[variable].rolling(window =window).mean()# The bigger the time window the smother the graph
    residual = data[variable]-avg                       # When I increase the value of sigma it makes outliers more, what plot
    std = residual.rolling(window = window).std()
    outliers = (np.abs(residual)>std*sigma)
    return view_fn(avg, avg[outliers])
```

```python
find_outliers(variable = 'Temperature', window = 20, sigma =10) # The lower the sigma the wider the rang to get the outliers from...
```

### Making data outliers for 'Temperature' interactive

```python
import panel as pn
pn.extension()
pn.interact(find_outliers)
```

## Making data outliers for all variables at once interactive

```python
kw = dict(window=(1,60), sigma=(1,20), variable=sorted (list(data.columns)))
calling_panal = pn.interact(find_outliers, **kw)
print(calling_panal)
```

```python
import param

class RoomOccupancy(param.Parameterized):
    variable  = param.Selector(objects=list(data.columns))
    window    = param.Integer(default=10, bounds=(1, 20))
    sigma     = param.Number(default=10, bounds=(0, 20))

    def view(self):
        return find_outliers(self.variable, self.window, self.sigma)
    
obj = RoomOccupancy()
obj
```

```python
pn.Row(obj.param, obj.view)
```

```python
import hvplot.pandas

def hvplot(avg, highlight):
    return avg.hvplot(height=200) * highlight.hvplot.scatter(color='orange', padding=0.1)

text2 = "## Room Occupancy\nSelect the variable and the smoothing values"
hvp   = pn.interact(find_outliers, view_fn=hvplot, **kw)
pn.Column(pn.Row(pn.panel(text2, width=400), hvp[0]), hvp[1]).servable("Occupancy")
```

```python
import holoviews as hv

tap = hv.streams.PointerX(x=data.index.min())

def hvplot2(avg, highlight):
    line = avg.hvplot(height=300, width=500)
    outliers = highlight.hvplot.scatter(color='orange', padding=0.1)
    tap.source = line
    return (line * outliers).opts(legend_position='top_right')

@pn.depends(tap.param.x)
def table(x):
    index = np.abs((data.index - x).astype(int)).argmin()
    return data.iloc[index]

app = pn.interact(find_outliers, view_fn=hvplot2, **kw)

pn.Row(
    pn.Column("## Room Occupancy\nHover over the plot for more information.", app[0]),
    pn.Row(app[1], table)
)
```
